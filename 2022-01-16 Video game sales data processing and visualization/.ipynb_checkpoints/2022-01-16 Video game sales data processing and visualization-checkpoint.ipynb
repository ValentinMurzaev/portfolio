## Video game sales data processing and visualization (using Pandas, Matplotlib, Seaborn)

The `vgsales.csv` dataset (will be uploaded below) contains information about the total sales of video games for video game consoles. Description of the features:

* `Rank` – the serial number of the record in the dataset
* `Name` – the name of the game
* `Platform` – the platform for which the game was released (Nintendo, PlayStation, XBox, etc.)
* `Year` – the year of release of the game
* `Genre` – the genre of the game
* `Publisher` – the name of the game's publishing company
* `NA_Sales` – sales volume of the game in North America, million copies
* `EU_Sales` – sales volume of the game in Europe, million copies
* `JP_Sales` – sales volume of the game in Japan, million copies
* `Other_Sales` – sales volume of the game in other countries, million copies
* `Global_Sales` – the volume of sales of the game worldwide, million copies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Loaded dataframe:

```python
df = pd.read_csv('vgsales.csv')
df
```

A histogram of the `sales volume distribution` for all the games in `Japan`:

```python
series = df['JP_Sales']
bin_width = series.max() / 10
x_range = [round(i * bin_width) for i in range(11)]

plt.hist(series, bins=50)
plt.title('Sales volume of all the games in Japan, 1980-2020')
plt.xlabel('M copies')
plt.ylabel('number of records')
plt.xticks(x_range)
plt.show()
```

A chart showing a comparison of global sales of the game `Grand Theft Auto V` on different platforms:

```python
series1 = df[df['Name'] == 'Grand Theft Auto V']

plt.bar(series1['Platform'], series1['Global_Sales'])
plt.title('Global sales of the Grand Theft Auto V game on different platforms')
plt.xlabel('Platform')
plt.ylabel('M copies')
plt.show()
```

A graph of the dynamics of total global sales of all games by years:

```python
df1 = df.groupby('Year').agg('sum')

plt.plot(df1['Global_Sales'])
plt.title('All games total global sales dynamics')
plt.xlabel('year')
plt.ylabel('M copies')
plt.show()
```

A chart showing the existence of worldwide sales dependence in North America, aggregated by position:

```python
plt.scatter(df['NA_Sales'], df['Global_Sales'])
plt.title('Worldwide different games sales depedence from N. America sales')
plt.xlabel('sales in N. America, M copies')
plt.ylabel('global sales, M copies')
plt.show()
```

The same applies to sales aggregated by years:

```python
plt.scatter(df1['NA_Sales'], df1['Global_Sales'])
plt.title('Worldwide different games sales depedence\nfrom N. America sales')
plt.xlabel('sales in N. America, M copies')
plt.ylabel('global sales, M copies')
plt.show()
```

Chart of the percentage of sales of the game `Super Mario Bros.` on different platforms worldwide:

```python
series2 = df[df['Name'] == 'Super Mario Bros.'].groupby('Platform').agg('sum')['Global_Sales']
plt.pie(series2, labels=series2.index, autopct='%1.1f%%')
plt.title('The share of different platforms\nin sales of the Super Mario Bros. game')
plt.show()
```

The 5 publishers with the highest total sales Worldwide in 2013:

```python
series3 = df[df['Year'] == 2013].groupby('Publisher').agg('sum').sort_values('Global_Sales', ascending=False).head(5)

plt.bar(series3.index, series3['Global_Sales'])
plt.title('The 5 publishers with the highest total sales Worldwide in 2013')
plt.xlabel('publishers')
plt.ylabel('sales, M copies')
plt.xticks(rotation=60)
plt.show()
```

A histogram of the global sales distribution of the games released by non-Nintendo publisher in the period from 2000 to 2015 inclusive:

```python
series4 = df[(df['Year'] >= 2000) & (df['Year'] <= 2015) & (df['Publisher'] != 'Nintendo')]['Global_Sales']
bin_width = series4.max() / 20
x_range = [i * bin_width for i in range(21)]

plt.hist(series4, bins=20)
plt.title('A histogram of the global sales distribution of the games\nreleased by non-Nintendo publishe, 2000-2015')
plt.xlabel('M copies')
plt.ylabel('number of records')
plt.xticks(x_range, rotation=45)
plt.show()
```

A graph of the total global sales dynamics of `Action` genre games by years:

```python
series5 = df[df['Genre'] == 'Action'].groupby('Year').agg('sum')
plt.plot(series5['Global_Sales'])
plt.title('Genre Acrion games world sales')
plt.xlabel('year')
plt.ylabel('M copies')
plt.show()

```

A chart illustrating the sales volume of all `Microsoft Game Studios` games in North America over time, divided by the genre:

```python
series6 = df[df['Publisher'] == 'Microsoft Game Studios'].groupby('Genre').agg('sum').\
sort_values('NA_Sales', ascending=False)

plt.bar(series6.index, series6['NA_Sales'])
plt.title("Sale volume of all Microsoft Game Studios' games\nin N. America")
plt.xlabel('genres')
plt.ylabel('sales, M copies')
plt.xticks(rotation=90)
plt.show()
```

A chart of the sales volume of all games of the `Sports` genre in Japan:

```python
fig, ax = plt.subplots(figsize=(12, 6))

data = df[df['Genre'] == 'Sports'][['Year', 'JP_Sales']].groupby('Year').agg('sum')

ax.plot(data.index, data['JP_Sales'])
ax.set_title('The genre Sports games sales dynamics in Japan')
ax.set_xlabel('year')
ax.set_ylabel('sales, M copies')

plt.show()
```

`Activision` Studio sales charts in North America, Europe, Japan and worldwide:

```python
fig, ax = plt.subplots(figsize=(12, 6))

data = df[df['Publisher'] == 'Activision'][['Year', 'NA_Sales']].groupby('Year').agg('sum')
data1 = df[df['Publisher'] == 'Activision'][['Year', 'EU_Sales']].groupby('Year').agg('sum')
data2 = df[df['Publisher'] == 'Activision'][['Year', 'JP_Sales']].groupby('Year').agg('sum')

ax.plot(data.index, data['NA_Sales'], label='N. America')
ax.plot(data.index, data1['EU_Sales'], label='Europe')
ax.plot(data.index, data2['JP_Sales'], label='Japan')
ax.set_title('Studio Activision games sales dymamics in different regions')
ax.set_xlabel('year')
ax.set_ylabel('sales, M copies')
ax.legend()

plt.show()
```

The same thing, but presented whth the separate charts:

```python
fig, axs = plt.subplots(nrows=3, sharex=True, figsize=(12, 6))

reg_sales = ['NA_Sales', 'EU_Sales', 'JP_Sales']
regs = ['N. America', 'Europe', 'Japan']

for i, reg_sale in enumerate(reg_sales):
    data = df[df['Publisher'] == 'Activision'][['Year', reg_sale]].groupby('Year').agg('sum')
    ax = axs[i]
    ax.plot(data.index, data[reg_sale])
    ax.set_title(regs[i], y=0.97)
    ax.set_ylabel('M copies')

fig.suptitle('Studio Activision games sales dymamics in different regions', y=0.98)

plt.show()
```

Another way to present this information:

```python
fig, axs = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(12, 6))

reg_sales = [['NA_Sales', 'EU_Sales'], ['JP_Sales', 'Global_Sales']]
regs = [['N. America', 'Europe'], ['Japan', 'World']]

for i in range(2):
    for j in range(2):
        data = df[df['Publisher'] == 'Activision'][['Year', reg_sales[i][j]]].groupby('Year').agg('sum')
        ax = axs[i][j]
        ax.plot(data.index, data[reg_sales[i][j]])
        ax.set_title(regs[i][j])
        if j == 0:
            ax.set_ylabel('M copies')

fig.suptitle('Studio Activision games sales dymamics in different regions', y=0.98)

plt.show()
```

Histograms of the global sales distribution of games published by `Microsoft Game Studios` and `Take-Two Interactive` after 2010 (2010 included):

```python
data = df[(df['Year'] >= 2000)]['Global_Sales']
data1 = data[(df['Publisher'] == 'Microsoft Game Studios')]
data2 = data[(df['Publisher'] == 'Take-Two Interactive')]

fig, ax = plt.subplots(figsize=(12, 6))
_, bins1, _ = ax.hist(data1, label='Microsoft Game Studios', alpha=0.5, bins=20)
ax.hist(data2, label='Take-Two Interactive', alpha=0.5, bins=bins1)

ax.set_title('A distribution of global sales of games from different publishers\nsince 2010')
ax.set_xlabel('annual sales of specific games, M copies')
ax.set_ylabel('number of records')
ax.set_xticks(list(bins1))
ax.legend()

plt.show()
```

The same thing created by `Seaborn` tools:

```python
data3 = df[(df['Publisher'] == 'Microsoft Game Studios') | (df['Publisher'] == 'Take-Two Interactive')
           & (df['Year'] >= 2010)]
fig, ax = plt.subplots(figsize=(12,6))
ax = sns.histplot(data=data3, x='Global_Sales', hue='Publisher', alpha=0.5, binwidth=1)
ax.set_title('A distribution of global sales of games of 2 publishers\nsince 2010')
ax.set_xlabel('sales, M copies')
ax.set_ylabel('number of records')
plt.show()
```

The 5 platforms for which the most games have been sold in Japan for all the monitoring time:

```python
data = df.groupby('Platform').agg('sum').sort_values('JP_Sales', ascending=False).head(5)
data['JP_Sales']
```

Visualization of this data on a bar chart, with the maximum and minimum values highlighted in separate colors:

```python
data['Color'] = 'b'
data.loc[data['JP_Sales'] == data['JP_Sales'].max(), 'Color'] = 'g'
data.loc[data['JP_Sales'] == data['JP_Sales'].min(), 'Color'] = 'r'

fig, ax = plt.subplots(figsize=(6, 3))
ax.bar(data.index, data['JP_Sales'], color=data['Color'])

ax.set_title('Sales of leading gaming platforms in Japan')
ax.set_xlabel('platforms')
ax.set_ylabel('M copies')

plt.show()
```

PairPlot graph for all tge numerical features of the dataset:

```python
data2 = df[['Platform', 'NA_Sales', 'EU_Sales', 'Other_Sales', 'Global_Sales']]
sns.pairplot(data2, hue='Platform')
plt.show()
```

A graph of the dynamics of the number of games released by `Nintendo`, by years, highlighting the interval in which the publisher released more than 35 games each year:

```python
data4 = df[df['Publisher'] == 'Nintendo'].groupby('Year')
game_n = pd.DataFrame(data4.agg('size'))
game_n = game_n.index[game_n[0] > 35]

fig, ax = plt.subplots(figsize=(12,6))
ax = sns.lineplot(data=data4.agg('sum'), x='Year', y='Global_Sales')
ax.set_title('Dynamics of global sales of all the games released by Nintendo')
ax.set_xlabel('year')
ax.set_ylabel('M copies')

ax.axvspan(xmin=game_n.min(), xmax=game_n.max(), color='g', alpha=0.5)

plt.show()
```

The 3 genres with the biggest game sales in the World for all the monitoring time:

```python
best_genre = df.groupby('Genre').agg('sum').sort_values('Global_Sales', ascending=False).head(3).index
best_genre
```

The 4 platforms with the biggest game sales in the World for all the monitoring time:

```python
best_platf = df.groupby('Platform').agg('sum').sort_values('Global_Sales', ascending=False).head(4).index
best_platf
```

A 4 by 3 graphs grid, in each cell of which there is a dot graph of the dependence of global sales on sales in North America for each pair (genre, platform):

```python
data5 = df[(df['Genre'].isin(best_genre)) & (df['Platform'].isin(best_platf))]

g = sns.FacetGrid(data5, row='Platform', col='Genre', hue='Year', palette='rocket', height=4)
g.map(sns.scatterplot, 'Global_Sales', 'NA_Sales')
g.add_legend()

plt.show()
```
